```python
# Установка функций СОЗДАНИЕ DF ПО ДИСЦИПЛИНЕ

sch_comm = pd.read_excel(r'\\192.168.128.12\VTBN_Disk\% ДПС\02 - ОРЦ\05 - ПИР\03 - ПРОЕКТ\03 - ПД\Rev\20_BIM\0_Schedules\ORC_P_I_X_SPH_BIM_Schedules_Комментарии ВТБН.xlsx')
sch_comm = sch_comm.drop(['Discipline','Category'],axis=1)

def ex_to_df_cnc(pr, nb, ds, block = ''):
    global partname5
    partname5 = ""
    
#pr -project (ORC or ORC), nb - number of building (1, 2, 7.4 ...), ds - discipline (AR, KR, OV, VK ...), cat - category (door, wall)
#    nsheet = take_sheet(cat)
    pathfile = take_path(pr, nb, ds, block)
    
    xls = xlrd.open_workbook(pathfile, on_demand=True)
    listsht = xls.sheet_names()
    listdf = xls.sheet_names()
    leng = len(listdf)  

    #Заполнение списка датафремов 
    for i in range(leng): 
        listdf[i] = pd.read_excel(pathfile, sheet_name = listdf[i])

    #Заполнение нулевых спецификаций
    for i in range(leng): 
        if (listdf[i].size == 0):
            listdf[i] = pd.DataFrame(columns=['Family and Type', 'Count'])

    #Установка мульти категорий           
    df_mc = pd.read_excel(pathfile, sheet_name = 'VTBR_Multi-Category')
    df_mcmt = pd.read_excel(pathfile, sheet_name = 'VTBR_Multi-Category MT')
    if (df_mcmt.empty == True):
        df_mcmt = df_mc
    #Объединение мульти и мульти материал
    df_m = pd.concat([df_mcmt,df_mc], join = "outer")
    #Установка датафр помещений  
    if 'VTBR_Room' in listsht:
        df_room = pd.read_excel(pathfile, sheet_name = 'VTBR_Room')

    #Удаление мульти спец (за хештегод - перемещение специв начало, в конец)  
    for i in range(len(listdf)):         
        if (listdf[i].equals(df_mcmt) == True):
            listdf.append(listdf[i])
            del listdf[i]
    for i in range(len(listdf)):           
        if (listdf[i].equals(df_mc) == True):
            listdf.append(listdf[i])
            del listdf[i]
    #listdf.insert(0 , listdf[-2])
    #listdf.insert(0 , listdf[-1])
    del listdf[-2:]


    #Удаление датафр помещений  
    if 'VTBR_Room' in listsht:    
        for i in range(len(listdf)):         
            if (listdf[i].equals(df_room) == True):
                listdf.append(listdf[i])
                del listdf[i]
        del listdf[-1:]
    
    for i in range(len(listdf)):          
        listdf[i].insert(0, 'Category', (listdf[i].merge(df_m, on = 'Family and Type', how = 'left').Category))
 
    
    df_cnc = pd.concat([listdf[0]
                           ], join = "outer")
    for i in range(1, len(listdf)):
        df_cnc = pd.concat([df_cnc,
                            listdf[i]
                           ], join = "outer")
    #Заплатка для нулевой мултиматериал
    if (df_cnc.empty == True):
        df_cnc = pd.DataFrame(columns=['Family and Type', 'Count']) 
        
    df_cnc.reset_index(inplace=True, drop=True)
    df_cnc.insert(0, 'Discipline', ds)
   
    df_cnc.insert(6, 'Count1', df_cnc['Count'])
    df_cnc.drop(['Count'],axis=1, inplace=True)
    df_cnc.rename(columns = {"Count1" : "Count"}, inplace=True)
    #Объединение с таблице комментариев       
    df_cnc = df_cnc.merge(sch_comm, on = 'Family and Type', how = 'left')
       
    return df_cnc
```

```python
def compare_mcat(pr, nb, ds, block = ''):
    
    pathfile = take_path(pr, nb, ds, block)
    
    #Создание списка категорий из ексель файла
    xl = pd.ExcelFile(pathfile)
    xl_sheets = xl.sheet_names

    #Разделение имени по "_"
    for i in range(len(xl_sheets)):
        xl_sheets[i] = xl_sheets[i].split('_')
        del xl_sheets[i][:1]
        xl_sheets[i] = ''.join(xl_sheets[i])
    xl_sheets.remove('Multi-Category')
    xl_sheets.remove('Multi-Category MT')
    #Создание DF категорий из ексель файла      
    df_m_xl = pd.DataFrame(xl_sheets, columns =['Excel'])

    
    #Установка мульти категорий           
    df_mc = pd.read_excel(pathfile, sheet_name = 'VTBR_Multi-Category')
    df_mcmt = pd.read_excel(pathfile, sheet_name = 'VTBR_Multi-Category MT')
    #Объединение мульти и мульти материал
    df_m = pd.concat([df_mcmt,df_mc], join = "outer")
    df_m_gr = df_m.groupby('Category', as_index=False).aggregate({'Family and Type': 'count'}).drop('Family and Type', axis = 1)
    df_m_gr.rename(columns = {"Category" : "Multi_Schedule"}, inplace=True)
    df_m = df_m_gr.join(df_m_xl)
    df_m['Multi_Schedule'] = df_m.Multi_Schedule.sort_values(ascending=True)
    df_m['Excel'] = df_m.Excel.sort_values(ascending=True)
    #Добавление столбцов для проверки
    df_m['Compare_sch'] = ''
    lst_sort = df_m_xl.Excel.sort_values(ascending=True, ignore_index=True)
    df_m = df_m.assign(Excel_sort = pd.Series(lst_sort))
    df_m = df_m.assign(Excel_s = df_m.Excel + 's')
    df_m = df_m.assign(Excel_es = df_m.Excel + 'es')
    df_m = df_m.assign(Excel_curtp = 'Curtain ' + df_m.Excel + 's')
    # Создание проверочного списка
    lst_excel = df_m['Excel'].to_list() + df_m['Excel_s'].to_list() + df_m['Excel_es'].to_list() + df_m['Excel_curtp'].to_list()

    df_m['Compare_sch'] = df_m['Multi_Schedule'].isin(lst_excel)
    df_m = df_m.drop(['Excel', 'Excel_es','Excel_s', 'Excel_curtp'], axis = 1)
    df_m.insert(0, 'Discipline', ds)
    return df_m

```
